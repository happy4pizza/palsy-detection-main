```python
import pandas as pd
import re
from pathlib import Path
```

```python
DATA_DIR = Path('../data/')
RAW_DIR = DATA_DIR / 'raw'
XLSX_PATH = RAW_DIR / 'pat_info.xlsx'
```

```python
df_meta = pd.read_excel(XLSX_PATH)

print(df_meta.shape)
print(df_meta.columns)
df_meta.head()
```

```python
records = []

for path in RAW_DIR.rglob("*"):
    suffix = path.suffix.lower()

    if suffix not in {".jpg", ".mp4"}:
        continue

    filename = path.name
    patient_id = path.parent.name

    # modality + pose
    if suffix == ".mp4":
        modality = "video"
        pose_index = None
    else:
        modality = "image"
        match = re.search(r"_(\d+)\.jpg$", filename.lower())
        pose_index = int(match.group(1)) if match else None

    parent = path.parents[1].name
    grandparent = path.parents[2].name

    # handle Normal separately
    if parent == "Normal":
        cohort = "Normal"
        severity_folder = "Normal"

    else:
        cohort = grandparent
        severity_folder = parent

    records.append({
        "patient_id": patient_id,
        "cohort": cohort,
        "severity_folder": severity_folder,
        "modality": modality,
        "pose_index": pose_index,
        "filepath": str(path),
    })

df = pd.DataFrame(records)
```

```python
# number of patients
print("Unique patients:", df["patient_id"].nunique())

# images per patient
df.groupby("patient_id").size().describe()

# pose coverage
df["pose_index"].value_counts().sort_index()
```

```python
def extract_hb(severity_folder):
    if severity_folder.startswith("Normal"):
        return 1
    if "NearNormal" in severity_folder:
        return 2
    if "Mild" in severity_folder:
        return 3
    if "Moderate" in severity_folder:
        return 4
    if "Severe" in severity_folder:
        return 5
    if "Complete" in severity_folder:
        return 6
    return None

df['hb_grade_folder'] = df['severity_folder'].apply(extract_hb)

df['hb_grade_folder'].value_counts()
```

```python
df_meta["patient_id"] = (
    df_meta['Sub-category'].astype(str).str.strip() +
    df_meta['Category'].astype(str).str.strip() +
    df_meta['#'].astype(str).str.strip()
)
df_meta = df_meta.dropna(subset=['patient_id'])
df['hb_grade'] = df['hb_grade_folder']
```

```python
set_files = set(df['patient_id'])
set_excel = set(df_meta['patient_id'])

print("Only in files:", set_files - set_excel)
print("Only in excel:", set_excel - set_files)
```

```python
def clean_patient_id(x):
    return str(x).strip()
df["patient_id"] = df["patient_id"].apply(clean_patient_id)
df_meta["patient_id"] = df_meta["patient_id"].apply(clean_patient_id)
```

```python
df_merged = df.merge(
    df_meta,
    on="patient_id",
    how="left"
)
```

```python
print(df_merged[["Category", "Sub-category", "#", "Side", "Gender", "Age"]].isnull().sum())
```

```python
print(
    df_merged.groupby("cohort")[["Age", "Gender", "Side"]]
    .apply(lambda x: x.isnull().sum())
)
```

```python
matched_patients = df_merged.loc[df_merged["Age"].notnull(), "patient_id"].nunique()
total_patients = df["patient_id"].nunique()

print("Matched patients:", matched_patients)
print("Total patients:", total_patients)
```

```python
print(df["cohort"].value_counts())
```

```python
df_final = df_merged[
    [
        "patient_id",
        "cohort",
        "severity_folder",
        "pose_index",
        "modality",
        "filepath",
        "hb_grade_folder",
        "Side",
        "Gender",
        "Age",
    ]
].copy()

df_final = df_final.rename(columns={"hb_grade_folder": "hb_grade"})
```

```python
df_final.head()
df_final["hb_grade"].value_counts().sort_index()
df_final.groupby("patient_id").size().describe()
```

```python
df_final.groupby("patient_id").size().value_counts().sort_index()
```

```python
OUT_PARQUET = DATA_DIR / "manifests" / "image_manifest.parquet"
OUT_CSV = DATA_DIR / "manifests" / "image_manifest.csv"

OUT_PARQUET.parent.mkdir(parents=True, exist_ok=True)

df_final.to_parquet(OUT_PARQUET, index=False)
df_final.to_csv(OUT_CSV, index=False)
```
